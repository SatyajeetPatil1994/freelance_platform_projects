```python
# Importing the required libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Load the dataset
df = pd.read_csv('freelance_platform_projects.csv')
df.head()
```

```python
df.drop(['Duration', 'Client Job Title','Location','Budget','Client Registration Date'], axis=1, inplace=True)
df.head()
```

```python
# Encoding the categorical variables
le = LabelEncoder()
df["Title"] = le.fit_transform(df["Title"])
df["Category Name"] = le.fit_transform(df["Category Name"])
df["Experience"] = le.fit_transform(df["Experience"])
df["Sub Category Name"] = le.fit_transform(df["Sub Category Name"])
df["Currency"] = le.fit_transform(df["Currency"])
df["Freelancer Preferred From"] = le.fit_transform(df["Freelancer Preferred From"])
df["Type"] = le.fit_transform(df["Type"])
df["Date Posted"] = le.fit_transform(df["Date Posted"])
df["Description"] = le.fit_transform(df["Description"])
df["Client City"] = le.fit_transform(df["Client City"])
df["Client Country"] = le.fit_transform(df["Client Country"])
df["Client Currency"] = le.fit_transform(df["Client Currency"])
```

```python
df.head()
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Type"]), df["Type"], test_size=0.2, random_state=42)
```

```python
# Training the Random Forest classifier
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, y_train)
```

```python
# Making predictions on the testing set
y_pred = rfc.predict(X_test)
y_pred[:10]
```

```python
# Evaluating the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
